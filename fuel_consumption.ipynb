Build your own linear regression as we did in the lecture to calculate the coefficients and intercept that minimizes the lease square error when predicting mpg using disp. Use the following training parameter values for initial weight and bias -0.1, 30. And for learning rate 0.00001 and 100 for the number of iterations. What are the values for the coefficients and intercept

```python
# Solve Here
def predict_MPG(feature, weights, bias):
   prediction = weights*feature + bias
   return prediction
```

```python
w1,b1 = [-0.1, 30]
y1 = predict_MPG(df.disp, w1, b1)
```

```python
def cost_function(true, predicted):
    error = sum((true - predicted)**2)/len(true)
    return error
```

```python
print(cost_function(df.disp,y1))
```

```python
def update_weights(features, true, weights, bias, learning_rate):
    deriv_w = -2/len(true)*sum(features*(true-(weights*features+bias)))
    deriv_b = -2/len(true)*sum((true-(weights*features+bias)))
    weights = weights - learning_rate*deriv_w
    bias = bias - learning_rate*deriv_b
    return weights, bias
```

```python
def train(features, true, weight, bias, learning_rate,iters):
    cost_history=[]
    for i in range(iters):
        predictions=predict_MPG(features,weight,bias)
        cost = cost_function(true, predictions)
        cost_history.append(cost)
        weight , bias = update_weights(features, true, weight, bias, learning_rate)
    return weight , bias, cost_history
```

```python
df
```

```python
w,b, c = train(df.disp, df.mpg, -0.1, 30., 0.00001,1000)
print(w,b)
print(c)
```
